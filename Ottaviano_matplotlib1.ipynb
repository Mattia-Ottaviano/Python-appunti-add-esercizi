1. crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python
import pandas as pd

data = {
  'last_name': ['Boyer','Copeland', 'Daniel','Serrano', 'Terrell','Vargas'],
  'orders_number': [185, 50, 196,58,24,28],
  }

df = pd.DataFrame(data)
df
```

2. visualizzi un grafico a barre verticali per rappresentare i dati. Il risultato deve essere questo:

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'], df['orders_number'])
```

3. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'], df['orders_number'], color ='r')
ax.set_xlabel('Commessi')
ax.set_ylabel('Ordini')
fig.suptitle('Numeri di Ordini dai Vari Commessi')
```

4. Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df.sort_values('orders_number', ascending= False)['last_name'], df['orders_number'].sort_values(ascending= False),color='g')
```

5. visualizzi un grafico a barre orizzontali per rappresentare i dati. Il risultato deve essere uguale a questo:

```python
fig = plt.figure()
ax = plt.axes()
ax.barh(df['last_name'], df['orders_number'], color ='r')
ax.set_xlabel('Commessi')
ax.set_ylabel('Ordini')
fig.suptitle('Numeri di Ordini dai Vari Commessi')
```

7. visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python
cols = ['c', 'm', 'r','g','b','y']

fig = plt.figure(figsize=[7,7])
ax = plt.axes()
ax.pie(df['orders_number'], labels=df['last_name'], colors= cols, startangle=90, autopct = '%1.1f%%')
fig.suptitle('Grafico Percentuale Dipendenti', color='k')

plt.show()
```
