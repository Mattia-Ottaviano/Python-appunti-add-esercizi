```
# installare (se necessario) e importare contextily e visualizzare la mappa di Milano con l'indicazione dei quartieri in nero (solo confini)
# e le piste ciclabili
```

```
ax = quartieri_milano.to_crs(epsg=3857).plot(facecolor='none', edgecolor='k', figsize= (20,15), linewidth=3)
piste_ciclabili = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Geopandas esercizio 3/bike_ciclabili.zip').to_crs(epsg=3857).plot(facecolor='none', edgecolor='b', ax = ax, linewidth=2)

ctx.add_basemap(ax)
```

```
# trovare il quartiere meno esteso
```

```
q_meno_esteso = quartieri_milano[quartieri_milano['Shape_Area'] == min(quartieri_milano['Shape_Area'])]
q_meno_esteso
```

```
# trovare il quartiere più esteso
```

```
q_piu_esteso = quartieri_milano[quartieri_milano['Shape_Area'] == max(quartieri_milano['Shape_Area'])]
q_piu_esteso
```

```
# visualizzare i due quartieri sulla mappa di Milano con i colori indicati nella figura sotto
```

```
ax = quartieri_milano[quartieri_milano['Shape_Area'] == max(quartieri_milano['Shape_Area'])].to_crs(epsg=3857).plot(facecolor='yellow', alpha=0.75, figsize=(15,10))
quartieri_milano[quartieri_milano['Shape_Area'] == min(quartieri_milano['Shape_Area'])].to_crs(epsg=3857).plot(ax=ax, alpha=0.75, facecolor='r')
ctx.add_basemap(ax)
```

```
# salvare in un geodataframe le informazioni sul quartiere di Affori, prendendole dal geodataframe dei quartieri
```

```
q_affori = quartieri_milano.loc[quartieri_milano['NIL'].str.contains('AFFORI')]
q_affori
```

```
# trovare le piste ciclabili che passano per il quartiere di Affori
```

```
piste_ciclabili[piste_ciclabili.intersects(q_affori.geometry.squeeze())]
```

```
# visualizzare su una mappa il quartiere di Affori e le piste ciclabili che lo attraversano
# evidenziare bene in rosso le piste ciclabili come nell'esempio sotto
```

```
ax = piste_ciclabili[piste_ciclabili.intersects(q_affori.geometry.squeeze())].to_crs(epsg=3857).buffer(7).plot(facecolor='r', figsize=(20,15))
q_affori.to_crs(epsg=3857).plot(ax=ax, facecolor='none', edgecolor='k')
ctx.add_basemap(ax)
```
